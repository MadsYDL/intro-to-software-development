```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


W_values = [80, 76, 72, 68, 64, 60, 56, 52, 48, 44, 40]

t_140 = [0,   50, 101, 154, 210, 271, 340, 418, 506, 610, 732]
t_120 = [0,   52, 106, 165, 229, 300, 379, 469, 572, 693, np.nan]
t_100 = [0,   53, 112, 177, 249, 329, 418, 519, 636, 771, np.nan]

data = {
    "RPM":   [140]*len(W_values) + [120]*len(W_values) + [100]*len(W_values),
    "W_kip": W_values * 3,
    "t_s":   t_140 + t_120 + t_100             # time in seconds
}

df = pd.DataFrame(data)

#compute ROP
df["ROP_ft_hr"] = 3600 / df["t_s"].replace(0, np.nan)

print("Dataframe with ROP:")
print(df)

def analyze_rpm(df, rpm, threshold_fraction=0.05):


    sub = df[(df["RPM"] == rpm) & df["ROP_ft_hr"].notna()].copy()

    max_rop = sub["ROP_ft_hr"].max()

    # threshold WOB: first point where ROP > fraction of max ROP
    thresh_row = sub[sub["ROP_ft_hr"] > threshold_fraction * max_rop].iloc[0]
    threshold_wob = thresh_row["W_kip"]
    threshold_rop = thresh_row["ROP_ft_hr"]

    # best WOB
    max_row = sub.loc[sub["ROP_ft_hr"].idxmax()]
    best_wob = max_row["W_kip"]
    best_rop = max_row["ROP_ft_hr"]

    return threshold_wob, threshold_rop, best_wob, best_rop

for rpm in [100, 120, 140]:
    th_w, th_r, best_w, best_r = analyze_rpm(df, rpm)
    print(f"\nRPM = {rpm}")
    print(f"  Threshold WOB  ≈ {th_w:.0f} klb (ROP ≈ {th_r:.1f} ft/hr)")
    print(f"  Best WOB       ≈ {best_w:.0f} klb (max ROP ≈ {best_r:.1f} ft/hr)")


plt.figure(figsize=(8, 6))

for rpm in [100, 120, 140]:
    sub = df[df["RPM"] == rpm]
    plt.plot(sub["W_kip"], sub["ROP_ft_hr"], marker="o", label=f"{rpm} RPM")

plt.xlabel("WOB (1000 lb)")
plt.ylabel("ROP (ft/hr)")
plt.title("WOB vs ROP for Different RPM")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
